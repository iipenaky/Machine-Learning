```python
# Importing necessary libraries
import matplotlib.pyplot as plt

# Data for plot
size_of_plot = [250, 300, 150, 400, 500, 600, 350, 450, 200, 275, 325, 550, 450, 225, 375, 175, 275, 350, 500, 600]
land_price = [25000, 28500, 15750, 36000, 45500, 52200, 32000, 40500, 19000, 27250, 30000, 50000, 41500, 20750, 34500, 17500, 26500, 33000, 46000, 53500]

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(size_of_plot, land_price, color='blue', marker='o')

# Adding title and labels
plt.title('Scatter Plot of Land Price vs. Plot Size in Ghana')
plt.xlabel('Size of Plot (sq. meters)')
plt.ylabel('Land Price (GHS)')

# Display the plot
#plt.grid(false)
plt.show()

# Save the plot as an SVG file
plt.savefig('land_price_vs_plot_size.svg', format='svg')
```
